```python
import requests

def get_mondo_id(disease_name):
    if not disease_name:
        return {'error': 'Please provide a disease name'}
    
    # Query the Ontobee API for MONDO ID
    try:
        response = requests.get(f'http://www.ontobee.org/api/search?ontology=MONDO&keywords={disease_name}&exact=1')
        if response.status_code == 200:
            data = response.json()
            if data and 'entities' in data and data['entities']:
                mondo_id = data['entities'][0]['iri'].split('/')[-1]
                return {'disease_name': disease_name, 'mondo_id': mondo_id}
            else:
                return {'error': 'MONDO ID not found for the given disease name'}
        else:
            return {'error': 'Error querying Ontobee API'}
    except Exception as e:
        return {'error': str(e)}

if __name__ == '__main__':
    disease_name = input("Enter the disease name: ")
    result = get_mondo_id(disease_name)
    if 'error' in result:
        print(f"Error: {result['error']}")
    else:
        print(f"Disease Name: {result['disease_name']}")
        print(f"MONDO ID: {result['mondo_id']}")
```

```python
import requests

def fetch_assay_data_for_gene(gene):
    # Define the ChEMBL API endpoint
    api_url = "https://www.ebi.ac.uk/chembl/api/data/assay"
    
    # Define query parameters
    params = {
        "assay_type": "B",
        "description__icontains": gene
    }

    try:
        # Make a GET request to the API
        response = requests.get(api_url, params=params)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Return the JSON data
            return response.json()
        else:
            # Handle other status codes if needed
            print(f"Error: API request returned status code {response.status_code}")
            return None
    except requests.exceptions.RequestException as e:
        # Handle connection errors
        print(f"Error connecting to API: {e}")
        return None

# Example usage:
gene_value = "KRas"
assay_data = fetch_assay_data_for_gene(gene_value)
print(assay_data)
```

```python
import requests
import xml.etree.ElementTree as ET

def get_chembl_id(target_name):
    base_url = "https://www.ebi.ac.uk/chembl/api/data/target/search"
    params = {
        "q": target_name
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes

        root = ET.fromstring(response.content)
        targets = root.findall('.//target_chembl_id')

        if targets:
            return targets[0].text
        else:
            return "Target not found in ChEMBL."

    except requests.exceptions.RequestException as e:
        return f"Error: {e}"

# Example usage:
target_name = input("Enter the target name: ")
chembl_id = get_chembl_id(target_name)
print(f"ChEMBL ID for {target_name}: {chembl_id}")
```

```python
import requests
import xml.etree.ElementTree as ET

def get_ligand_ids(chembl_id):
    base_url = "https://www.ebi.ac.uk/chembl/api/data/activity"
    params = {
        "target_chembl_id__in": chembl_id,
        "assay_type": "B"
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes

        root = ET.fromstring(response.content)
        activities = root.findall('.//activity')

        ligand_ids = []
        for activity in activities:
            molecule_chembl_id = activity.find('molecule_chembl_id')
            if molecule_chembl_id is not None:
                ligand_ids.append(molecule_chembl_id.text)

        if ligand_ids:
            return ligand_ids
        else:
            return "No ligands found for the given ChEMBL ID."

    except requests.exceptions.RequestException as e:
        return f"Error: {e}"

# Example usage:
chembl_id = input("Enter the ChEMBL ID: ")
ligand_ids = get_ligand_ids(chembl_id)
print(f"Ligand IDs for {chembl_id}: {ligand_ids}")
```

```python
import requests
import xml.etree.ElementTree as ET

def get_chembl_ids(target_name):
    base_url = "https://www.ebi.ac.uk/chembl/api/data/target/search"
    params = {
        "q": target_name
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes

        root = ET.fromstring(response.content)
        targets = root.findall('.//target_chembl_id')

        chembl_ids = [target.text for target in targets]

        if chembl_ids:
            return chembl_ids
        else:
            return ["Target not found in ChEMBL."]

    except requests.exceptions.RequestException as e:
        return [f"Error: {e}"]

def get_ligand_chembl_ids(chembl_id):
    base_url = "https://www.ebi.ac.uk/chembl/api/data/activity"
    params = {
        "target_chembl_id__in": chembl_id,
        "assay_type": "B"
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes

        root = ET.fromstring(response.content)
        ligands = root.findall('.//molecule_chembl_id')

        ligand_chembl_ids = [ligand.text for ligand in ligands]

        if ligand_chembl_ids:
            return ligand_chembl_ids
        else:
            return ["No ligands found for this ChEMBL ID."]

    except requests.exceptions.RequestException as e:
        return [f"Error: {e}"]

def fetch_ligands_for_target(target_name):
    chembl_ids = get_chembl_ids(target_name)
    if not chembl_ids or any("Error" in cid for cid in chembl_ids) or "not found" in chembl_ids:
        return "Cannot fetch ligands without a valid ChEMBL ID."

    unique_ligand_ids = set()
    for chembl_id in chembl_ids:
        ligand_ids = get_ligand_chembl_ids(chembl_id)
        unique_ligand_ids.update(ligand_ids)

    return list(unique_ligand_ids)

# Example usage:
if __name__ == "__main__":
    target_name = input("Enter the target name: ")
    ligand_chembl_ids = fetch_ligands_for_target(target_name)
    print(f"Ligand ChEMBL IDs for {target_name}: {ligand_chembl_ids}")
```
